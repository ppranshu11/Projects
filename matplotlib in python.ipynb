```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
# pie chart in matplotlib
```

```python
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,25,25,15])
my_label=["apple","banana","cherry","grapes"]

plt.pie(y,labels=my_label)
plt.show()
```

```python
y=np.array([35,25,25,15])
my_label=["apple","banana","cherry","grapes"]

plt.pie(y,labels=my_label,startangle=90)
plt.show()
```

```python
y=np.array([35,25,25,15])
my_label=["apple","banana","cherry","grapes"]

my_explode=[0.2,0.5,0.7,1.0]

plt.pie(y,labels=my_label,explode=my_explode)

plt.show()
```

```python
y=np.array([35,25,25,15])
my_label=["apple","banana","cherry","grapes"]

my_explode=[0.2,0,0,0]

plt.pie(y,labels=my_label,explode=my_explode,shadow=True)

plt.show()
```

```python
y=np.array([35,25,25,15])
my_label=["apple","banana","cherry","grapes"]
my_color=["red","yellow","hotpink","green"]

plt.pie(y,labels=my_label,colors=my_color)
plt.show()
```

```python
#legend
```

```python
y=np.array([35,25,25,15])
my_label=["apple","banana","cherry","grapes"]

plt.pie(y,labels=my_label)
plt.legend()
plt.show()
```

```python
#legend with header
```

```python
y=np.array([35,25,25,15])
my_label=["apple","banana","cherry","grapes"]

plt.pie(y)

plt.legend(labels=my_label,title="fruits")
plt.show()
```
