```python
# wrong data
df=pd.read_csv(r"C:\Users\rahulpathak\Downloads/Health_data.csv")
print(df.to_string())
```

```python
df.loc[4,"Duration"]=45
print(df.to_string())
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.loc[x,"Duration"]=120
print(df.to_string())
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.drop(x,inplace=True)
print(df.to_string())
```

```python
df.loc[10,"Pulse"]=100
print(df.to_string())
```

```python
# removing duplicates
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
print(df.to_string())
```
